```python
import yfinance as yf
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Disney = yf.Ticker("DIS")
Disney = yf.download('DIS','2021-10-05','2021-11-19')

SP500 = yf.Ticker("^GSPC")
SP500 = yf.download('^GSPC','2021-10-05','2021-11-19')

DOW = yf.Ticker("^DJI")
DOW = yf.download('^DJI','2021-10-05','2021-11-19')
```

```python
import pandas as pd

stock = pd.DataFrame({'Disney':Disney["Adj Close"],'SP500':SP500["Adj Close"],'DOW':DOW["Adj Close"]})
stock
```

```python
plot = stock.plot(figsize=(16, 9))
```

```python
stock.to_csv("Eternals.csv")
```

```python
fig = plot.get_figure()
fig.savefig('Raya and the Last Dragon.png')
```
